```python
import requests
# Clean the link column by removing the unwanted information
unique_classes_df['link'] = unique_classes_df['link'].str.rsplit('/#L', n=1).str[0]

# Add a new column 'code_snippet' to the DataFrame
unique_classes_df['code_snippet'] = ''

# Iterate over the rows in the DataFrame
for index, row in unique_classes_df.iterrows():
    link = row['link']
    start_line = row['start_line']
    end_line = row['end_line']

    # Send a GET request to retrieve the raw file content from GitHub
    raw_url = link.replace("github.com", "raw.githubusercontent.com").replace("/blob/", "/")
    response = requests.get(raw_url)

    # Check if the request was successful
    if response.status_code == 200:
        # Split the response content into lines
        lines = response.text.split("\n")

        # Extract the lines specified by start_line and end_line
        snippet_lines = lines[start_line - 1:end_line]

        # Join the code snippet lines into a single string
        snippet = "\n".join(snippet_lines)

        # Assign the snippet to the 'code_snippet' column
        unique_classes_df.at[index, 'code_snippet'] = snippet
    else:
        print(f"Failed to retrieve code from link: {link}")

# Print the updated DataFrame
print(unique_classes_df)
```

```python


def calculate_cyclomatic_complexity(code):
    try:
        # Tokenize and parse the evaluated code
        tokens = javalang.tokenizer.tokenize(code)
        parser = javalang.parser.Parser(tokens)
        tree = parser.parse_member_declaration()

        # Count the number of decision points (branches and loops)
        complexity = 1  # Start with a complexity of 1 for the method itself
        for _, node in tree:
            if isinstance(node, javalang.tree.IfStatement):
                complexity += 1
            elif isinstance(node, javalang.tree.ForStatement):
                complexity += 1
            elif isinstance(node, javalang.tree.WhileStatement):
                complexity += 1
            elif isinstance(node, javalang.tree.DoStatement):
                complexity += 1
            elif isinstance(node, javalang.tree.SwitchStatement):
                complexity += len(node.cases)

        return complexity

    except (SyntaxError, ValueError) as e:
        print("Error evaluating code:")
        print(code)
        raise e

# Create a new column 'cyclomatic_complexity' in the DataFrame 'df'
parsed_unique_classes_df['cyclomatic_complexity'] = parsed_unique_classes_df['code_snippet'].apply(calculate_cyclomatic_complexity)
display(parsed_unique_classes_df)
```

```python
import re
def count_max_method_chain_length(code):

  # Remove comments (optional, adjust pattern if needed)
  code = re.sub(r"//.*|\/\*.*?\*\/", "", code, flags=re.DOTALL)

  # Split by statements (semicolons)
  statements = code.split(";")

  # Initialize maximum chain length
  max_chain_length = 0

  for statement in statements:
    # Count dot-separated method calls (basic pattern)
    chain_length = len(re.findall(r"\.[a-zA-Z0-9_]+\(", statement))
    max_chain_length = max(max_chain_length, chain_length)

  return max_chain_length





# Apply the function to the 'code_snippet' column of your DataFrame
parsed_unique_classes_df['method_chain_length'] = parsed_unique_classes_df['code_snippet'].apply(count_max_method_chain_length)

print(f"Added 'estimated_mcl' column to DataFrame with estimated maximum chain lengths.")
```

```python
def calculate_atfd(code):
    # Tokenize and parse the evaluated code
    tokens = javalang.tokenizer.tokenize(code)
    parser = javalang.parser.Parser(tokens)
    tree = parser.parse_member_declaration()

    atfd = 0  # Initialize the ATFD metric to 0

    field_declarations = set()  # Store the field declarations in the class

    for path, node in tree:
        if isinstance(node, javalang.tree.FieldDeclaration):
            # Add the field names to the set of field declarations
            for declarator in node.declarators:
                field_declarations.add(declarator.name)

        if isinstance(node, javalang.tree.MethodDeclaration):
            for parameter in node.parameters:
                # Check if the parameter type is from an unrelated class
                if parameter.type.name not in field_declarations:
                    atfd += 1  # Increment the ATFD metric

            if node.body is not None:
                for statement in node.body:
                    # Check if the statement is a method invocation
                    if isinstance(statement, javalang.tree.MethodInvocation):
                        # Check if the invoked method belongs to an unrelated class
                        if statement.member not in field_declarations:
                            atfd += 1  # Increment the ATFD metric

    return atfd

# Create a new column 'atfd' in the DataFrame 'parsed_unique_god_class_df'
parsed_unique_classes_df['atfd'] = parsed_unique_classes_df['code_snippet'].apply(calculate_atfd)
display(parsed_unique_classes_df)

```

```python
# The implementation of rfc here a little differ from
# the implementation of rfc in god class metric based but
# the same output
def calculate_rfc(code):
    # Tokenize and parse the evaluated code
    tokens = javalang.tokenizer.tokenize(code)
    parser = javalang.parser.Parser(tokens)
    tree = parser.parse_member_declaration()

    rfc = 0  # Initialize the RFC metric to 0

    method_names = set()  # Store the names of methods in the class

    for path, node in tree:
        if isinstance(node, javalang.tree.MethodDeclaration):
            # Add the method name to the set of method names
            method_names.add(node.name)

            if node.body is not None:
                for statement in node.body:
                    # Check if the statement is a method invocation
                    if isinstance(statement, javalang.tree.MethodInvocation):
                        # Add the invoked method name to the set of method names
                        method_names.add(statement.member)

    # Calculate RFC as the sum of the number of methods and methods invoked
    rfc = len(method_names)

    return rfc

# Create a new column 'rfc' in the DataFrame 'parsed_unique_classes_df'
parsed_unique_classes_df['rfc'] = parsed_unique_classes_df['code_snippet'].apply(calculate_rfc)
display(parsed_unique_classes_df)
```

```python
# The implementation of cbo here differ from
# the implementation of cbo in god class metric based
# here seems more acurate
def calculate_cbo(code):
    # Tokenize and parse the evaluated code
    tokens = javalang.tokenizer.tokenize(code)
    parser = javalang.parser.Parser(tokens)
    tree = parser.parse_member_declaration()

    cbo = 0  # Initialize the CBO metric to 0

    referenced_classes = set()  # Store the names of classes referenced by the class

    for path, node in tree:
        # Check if the node is a ClassDeclaration
        if isinstance(node, javalang.tree.ClassDeclaration):
            # Iterate through the class members
            for member in node.body:
                # Check if the member is a FieldDeclaration
                if isinstance(member, javalang.tree.FieldDeclaration):
                    # Add the type name of the field to the set of referenced classes
                    referenced_classes.add(member.type.name)

                # Check if the member is a MethodDeclaration
                elif isinstance(member, javalang.tree.MethodDeclaration):
                    # Iterate through the method body statements
                    if member.body is not None:
                        for statement in member.body:
                            # Check if the statement is a MemberReference (field or method access)
                            if isinstance(statement, javalang.tree.MemberReference):
                                # Add the type name of the referenced member to the set of referenced classes
                                referenced_classes.add(statement.member)

    # Calculate CBO as the number of unique classes referenced
    cbo = len(referenced_classes)

    return cbo

# Create a new column 'cbo' in the DataFrame 'parsed_unique_classes_df'
parsed_unique_classes_df['cbo'] = parsed_unique_classes_df['code_snippet'].apply(calculate_cbo)
display(parsed_unique_classes_df)
```

```python
import javalang

def count_external_method_calls(code):
    # Tokenize and parse the evaluated code
    tokens = javalang.tokenizer.tokenize(code)
    parser = javalang.parser.Parser(tokens)
    tree = parser.parse_member_declaration()

    external_method_call_count = 0  # Initialize the count of external method calls to 0

    # Set to store method names defined in the current class or its ancestors
    local_methods = set()

    # Traverse the tree to find method declarations in the current class or its ancestors
    for path, node in tree:
        if isinstance(node, javalang.tree.MethodDeclaration):
            local_methods.add(node.name)

    # Traverse the tree again to find method invocations
    for path, node in tree:
        if isinstance(node, javalang.tree.MethodInvocation):
            # Check if the method invocation is external (not defined locally)
            if node.member not in local_methods:
                external_method_call_count += 1

    return external_method_call_count

# Usage:
parsed_unique_classes_df['external_method_call_count'] = parsed_unique_classes_df['code_snippet'].apply(count_external_method_calls)
display(parsed_unique_classes_df)
```

```python
def calculate_comment_ratio(code):
    total_lines = 0
    comment_lines = 0

    # Split the code into lines
    lines = code.split('\n')

    # Iterate over each line
    for line in lines:
        total_lines += 1
        # Check if the line is a comment or contains a comment
        if line.strip().startswith("//") or "/*" in line:
            comment_lines += 1

    # Calculate the comment ratio
    if total_lines == 0:
        return 0
    else:
        return comment_lines / total_lines


parsed_unique_classes_df['comment_ratio_count'] = parsed_unique_classes_df['code_snippet'].apply(calculate_comment_ratio)
display(parsed_unique_classes_df)



```
